```python
import random
import matplotlib.pyplot as plt
```

```python
def polyas_urn(iterations): #Creates a function that simulates the problem with an input "iterations"
    red_balls = 1 #Start with 1 red ball
    green_balls = 1 #Start with 1 green ball
    ratio = [] #Creates an empty list to store the changing ratio of red balls in the urn
    
    for i in range(iterations): 
        total_balls = red_balls + green_balls #Counts total number of balls in the urn
        
        red_ratio = red_balls / total_balls #Determines ratio of red balls
        
        x = random.random() #Generates a random variable between 0 and 1
        if x < red_ratio: #Checks if x is less than the ratio of red balls in the urn
            red_balls += 1 #If true, adds another red ball in the urn
        else:
            green_balls += 1 #If false, adds a green ball in the urn

        updated_ratio = red_balls / (red_balls + green_balls) #Determines updated ratio of balls following addition
        ratio.append(updated_ratio) #Adds the new ratio of red balls to the running list of ratios
        
    return ratio
```

```python
trials = 1000 #Number of trials

ratio_2000_trials = [] #Creates an empty list to store the ratio of the 2000th iteration for each trial
ratio_4000_trials = [] #Creates an empty list to store the ratio of the 4000th iteration for each trial

for i in range(trials):
    iterations = 4000 #Number of times we draw a ball in a single trial
    ratio = polyas_urn(iterations) #Stores list of the ratio of red balls for 4000 iterations in a single trial
        
    ratio_2000 = ratio[(len(ratio)//2)-1] #Finds the ratio of red balls after 2000 iterations
    ratio_4000 = ratio[iterations-1] #Finds the ratio of red balls after 4000 iterations
        
    ratio_2000_trials.append(ratio_2000) #Adds the ratio of red balls on the 2000th iteration in a list for all trials
    ratio_4000_trials.append(ratio_4000) #Adds the ratio of red balls on the 2000th iteration in a list for all trials
    
#print(ratio_2000_trials)
#print(ratio_4000_trials)
```

```python
plt.hist(ratio_2000_trials, color = "red")
plt.title("Polya's Urn Trials - Ratio of Red Balls (2000 Iterations)")
plt.show()
```

```python
plt.hist(ratio_4000_trials, color = "red")
plt.title("Polya's Urn Trials - Ratio of Red Balls (4000 Iterations)")
plt.show()
```
